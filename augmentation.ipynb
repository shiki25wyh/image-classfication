```python
import pathlib
dir_path = './train'
dir_path = pathlib.Path(dir_path)

cat = list(dir_path.glob('*cat*.jpg'))
dog = list(dir_path.glob('*dog*.jpg'))
```

```python
PIL.Image.open(str(cat[0])).resize((300,300))
```

```python
with tf.device("/cpu:0"):
    input_img = np.array(PIL.Image.open(str(cat[0])).resize((300,300)))
    input_img=np.expand_dims(input_img,axis=0)
    
    Flip_h = tf.keras.layers.experimental.preprocessing.RandomFlip("horizontal")
    Flip_v = tf.keras.layers.experimental.preprocessing.RandomFlip("vertical")
    Flip_h_img = Flip_h(input_img)
    Flip_v_img = Flip_v(input_img)
    
    Zoom = tf.keras.layers.experimental.preprocessing.RandomZoom(0.4)
    Zoom_img = Zoom(input_img)
    
    
    
    Rotation = tf.keras.layers.experimental.preprocessing.RandomRotation(0.4)
    Rotation_img = Rotation(input_img)
    
    
    
    Crop = tf.keras.layers.experimental.preprocessing.RandomCrop(256,256)
    Crop_img = Crop(input_img)
    
    
    
    Gaussian = tf.keras.layers.GaussianNoise(0.4)
    Gaussian_img = Gaussian(input_img)

```

```python
fig,axs = plt.subplots(1,7,figsize=(20,20))
axs[0].imshow(input_img[0])
axs[1].imshow(Flip_h_img[0])
axs[2].imshow(Flip_v_img[0])
axs[3].imshow(Zoom_img[0])
axs[4].imshow(Rotation_img[0])
axs[5].imshow(Crop_img[0])
axs[6].imshow(Gaussian_img[0])

axs[0].set_title('Origin')
axs[1].set_title('Horizontal Flip')
axs[2].set_title('Vertical Flip')
axs[3].set_title('Zoom')
axs[4].set_title('Rotation')
axs[5].set_title('Crop')
axs[6].set_title('Gaussion noise')

plt.show()
```
